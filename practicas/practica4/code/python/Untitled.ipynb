```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_context('talk')
```

```python
data_dir = '../source'
```

```python
Tc_theo = 2/np.log(1+np.sqrt(2))
decorr = 1000000
transit = 1000000
Ti = 1.80
Tf = 3.0
delta_T = 0.05
L_values = [16, 32, 64]#, 128]
columns = ['T', 'e', 'm', 'Cp', 'chi', 'm2', 'm4']
dfs = {}
for L in L_values:
    input_file_name = "L{}_Ti{:.2f}_Tf{:.2f}_deltaT{:.2f}_transit{}_decorr{}.dat".format(L, Ti, Tf, delta_T,
                                                                                        transit, decorr)
    full_input_file_name = os.path.join(data_dir, input_file_name)
    data = np.loadtxt(full_input_file_name)
    df = pd.DataFrame(data, columns=columns)
    #df = df.set_index('T')
    dfs[L] = df
    
```

```python
fig, ax = plt.subplots()
for L in L_values:
    df = dfs[L]
    ax.axvline(Tc_theo, color='k', linestyle='--')
    ax.plot(df['T'], df.m)
    #ax.plot(df['T'], df.m2)
    #ax.plot(df['T'], df.m4/df.m2**2)
    #ax.set_ylim(-100, 100)
    #ax.set_xlim(2, 2.5)
```
